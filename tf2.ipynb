```
from PIL import Image
import os, glob, sys, numpy 
from sklearn.model_selection import train_test_split
import tensorflow as tf
from google.colab import drive
drive.mount('/content/gdrive/')
import shutil

#img_dir='/content/gdrive/My Drive/images_dog'
#!ls "/content/gdrive/My Drive/images_dog/Chihuahua"
img_path_Pome = "/content/gdrive/My Drive/images_dog/Pomeranian_grown/*.jpg"
img_path_husky = "/content/gdrive/My Drive/images_dog/Husky_grown/*.jpg"
img_path_Pometest = "/content/gdrive/My Drive/image_test/Pomeranian_grown_test/*.jpg"
img_path_Huskytest = "/content/gdrive/My Drive/image_test/Husky_grown_test/*.jpg"
list_Pome = glob.glob(img_path_Pome)
list_husky = glob.glob(img_path_husky)
list_Pometest=glob.glob(img_path_Pometest)
list_Huskytest=glob.glob(img_path_Huskytest)

def pre_process_imgae(img_path, size = 28): #이미지 전처리 함수
    from tensorflow.keras.preprocessing import image
    img = image.load_img(img_path, target_size=(size,size)) # img load , 28*28 size
    img_array = image.img_to_array(img) # img to array
    img_array /= 255 # 0~255 값을 0~1로 정규화
    return img_array # shape = 28 28 3

# 포메 3 #허스키 4
train_data = [] 
train_lable = []
test_data = []
test_lable = []
for i in list_Pome:
  train_data.append(pre_process_imgae(i) )
  train_lable.append(3)

for i in list_husky:
  train_data.append(pre_process_imgae(i))
  train_lable.append(4)

for i in list_Pometest:
  test_data.append(pre_process_imgae(i) )
  test_lable.append(3)

for i in list_Huskytest:
  test_data.append(pre_process_imgae(i))
  test_lable.append(4)

train_data = numpy.array(train_data)
train_lable = numpy.array(train_lable)
test_data = numpy.array(test_data)
test_lable = numpy.array(test_lable)

model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28, 3)), # 입력층
    tf.keras.layers.Dense(50, activation='relu'), # 은닉층
    tf.keras.layers.Dropout(0.4), # (학습>테스트) -> 과적합으로 판단하여 -> 드롭아웃 추가 ->은닉층 개수, 드롭아웃 확률 조정 ->간격완화
    tf.keras.layers.Dense(10, activation='softmax') #출력층
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_data, train_lable, epochs=90)

test_loss, test_acc = model.evaluate(test_data, test_lable, verbose=2)
```
